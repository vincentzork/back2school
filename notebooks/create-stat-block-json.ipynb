```python
from back2school import demon

# List of demon stat blocks
demon_stat_blocks = [
    demon.demogorgon_custom_stat_block,
    demon.orcus_custom_stat_block,
    demon.grazzt_custom_stat_block,
    demon.baphomet_custom_stat_block,
    demon.yeenoghu_custom_stat_block,
    demon.zariel_custom_stat_block,
    demon.fraz_urbluu_custom_stat_block,
    demon.juiblex_custom_stat_block,
    demon.zuggtmoy_custom_stat_block,
]

# Convert list to a dictionary with names as keys
demon_stat_blocks_dict = {demon["name"]: demon for demon in demon_stat_blocks}

# Save the dictionary to a JSON file
with open(directories.data("demon_stat_blocks.json"), "w") as f:
    json.dump(demon_stat_blocks_dict, f, indent=4)
```
